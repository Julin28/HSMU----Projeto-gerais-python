```python
#importando pandas
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#baixando o csv
arq = pd.read_csv('/content/julho_py2.csv', sep=';')
```

```python
#verificando a base
arq.head(5)
```

```python
#OBJETIVO MÁXIMO: ELABORAR UM GRÁFICO
#DESCRIÇAO: o total de infrações por tipo, com o eixo x representando o tipo e o eixo y o total;
plt.figure(figsize=(15,15))
arq['grav_tipo'].value_counts(ascending=True).plot.bar(title='multas_por_infracao')
plt.xlabel('tipo_infracao')
plt.ylabel('total') 
```

```python
#o total de infrações por tipo de infrator, como condutor e proprietário.
plt.figure(figsize=(15,15))
arq['tipo_infrator'].value_counts(ascending=True).plot.bar(title='multas_por_infrator')
plt.xlabel('tipo_infrator')
plt.ylabel('total') 
```
